```python
## 假如我們不想把資料載到自己的電腦裡?
# 把連結填入
target_url = "https://raw.githubusercontent.com/vashineyu/slides_and_others/master/tutorial/examples/imagenet_urls_examples.txt"
```

```python
import requests
response = requests.get(target_url)
data = response.text

# 用 request 傳送回來的資料不會認得斷行符號
print(len(data))
data[0:100]
```

```python
# 找到換行符號，用該符號做字串分割後，把它拿掉
split_tag = '\n'
#\t是tab \n是換行
data = data.split(split_tag)
print(len(data))
data[0]
```

## 將 txt 轉成 pandas dataframe

```python
icon = '\t'
id_list = []
image_url_list = []
for i in data:
    teb_index = i.find(icon)
    id_ = i[:teb_index]
    image_url = i[teb_index+1:]
    id_list.append(id_)
    image_url_list.append(image_url)
```

```python
import pandas as pd

arrange_data = {'id' : id_list ,
       'image_url' : image_url_list}

df = pd.DataFrame(arrange_data)
df.head()
```

```python
for first_five in range(5):
    print(df.loc[first_five][1])
```
