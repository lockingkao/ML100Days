```python
app_train_select = app_train.select_dtypes(include=["int","float"])
app_train_select
```

```python
specific_col = str(input("please input the column you want to claculate:"))
print("the column you select is:%s"%specific_col)
try:
    tra_mean = app_train_select[specific_col].mean()
    tra_var = app_train_select[specific_col].var()
    tra_all = app_train_select[specific_col].describe()
    print("mean:%f"%tra_mean)
    print("variance:%f"%tra_var)
    print(tra_all)
except:
    print("your input is wrong!please reinput the correct columns!")
```
